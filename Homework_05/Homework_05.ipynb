# 第五题
对两个函数线型（Gauss 分布和 Lorentz 分布），设其一为 p(x)，另一为 F(x)， 用舍选法对 p(x) 抽样。将计算得到的归一化频数分布直方图与理论曲线 p(x) 进行比较， 讨论差异。讨论抽样效率。
$$
\text { Gauss: } \frac{1}{\sqrt{2 \pi}} \exp \left(-\frac{x^{2}}{2}\right);
\text { Lorentz: } \frac{1}{\pi} \frac{1}{1+x^{2}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

# Plot the curves

```python
lin_x = np.linspace(-5, 5, 1000)
gauss_y = np.exp(-np.square(lin_x) / 2) / (np.sqrt(2 * np.pi))
Lorentz_y = 1 / (np.pi * (1 + np.square(lin_x)))
Lorentz_amp_y = 1.6 / (np.pi * (1 + np.square(lin_x)))
 
plt.plot(lin_x, gauss_y)
plt.plot(lin_x, Lorentz_y)
plt.plot(lin_x, Lorentz_amp_y)
 
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Guassian', 'Lorentz', 'Lorentz_amplified'])
 
plt.show()
```

```python
eta = np.random.rand(500000)
x = np.tan(2 * np.arctan(5) * eta - np.arctan(5))
y = np.random.rand(500000)

ratio = (np.exp(-np.square(x) / 2) / (np.sqrt(2 * np.pi))) / (1.6 / (np.pi * (1 + np.square(x))))
index = y <= ratio
x = x[index]
plt.hist(x, bins=1000, alpha=0.7, density=True)
plt.plot(lin_x, gauss_y)
plt.show()
```
