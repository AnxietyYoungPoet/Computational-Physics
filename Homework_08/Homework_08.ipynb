# 第8题
自设若干个随机分布（相同或不同发布，它们有相同或不同的$\mu,\sigma^2$） 通过Monte Carlo模拟，验证中心极限定理成立（N =2、5、10）。

## $$p(x)=\exp(-x), x=\log(\frac{1}{\xi}), x\in[0, \infty],\mu=1,\sigma=1$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
lin_x = np.linspace(-3, 3, 1000)
gauss_y = np.exp(-np.square(lin_x) / 2) / (np.sqrt(2 * np.pi))

N = 2
rands = np.random.rand(100000, N)
x = np.log(1 / rands)
data = (np.mean(x, axis=1) - 1) / (1 / np.sqrt(N))
plt.hist(data, bins=1000, alpha=0.7, density=True)
plt.plot(lin_x, gauss_y)
plt.show()
```

```python
N = 5
rands = np.random.rand(100000, N)
x = np.log(1 / rands)
data = (np.mean(x, axis=1) - 1) / (1 / np.sqrt(N))
plt.hist(data, bins=1000, alpha=0.7, density=True)
plt.plot(lin_x, gauss_y)
plt.show()
```

```python
N = 10
rands = np.random.rand(100000, N)
x = np.log(1 / rands)
data = (np.mean(x, axis=1) - 1) / (1 / np.sqrt(N))
plt.hist(data, bins=1000, alpha=0.7, density=True)
plt.plot(lin_x, gauss_y)
plt.show()
```

```python
N = 100
rands = np.random.rand(100000, N)
x = np.log(1 / rands)
data = (np.mean(x, axis=1) - 1) / (1 / np.sqrt(N))
plt.hist(data, bins=500, alpha=0.7, density=True)
plt.plot(lin_x, gauss_y)
plt.show()
```

## $$
p(x)=\frac{1}{\pi\left(1+x^{2}\right)}, x=\sin \left[\pi\left(\xi-\frac{1}{2}\right)\right], x \in[-1,1], \mu=0, \sigma=\sqrt{\frac{1}{2}}
$$

```python
N = 2
rands = np.random.rand(100000, N)
x = np.sin(np.pi * (rands - 1 / 2))
data = (np.mean(x, axis=1)) / (np.sqrt(1 / 2) / np.sqrt(N))
plt.hist(data, bins=500, alpha=0.7, density=True)
plt.plot(lin_x, gauss_y)
plt.show()
```

```python
N = 5
rands = np.random.rand(100000, N)
x = np.sin(np.pi * (rands - 1 / 2))
data = (np.mean(x, axis=1)) / (np.sqrt(1 / 2) / np.sqrt(N))
plt.hist(data, bins=500, alpha=0.7, density=True)
plt.plot(lin_x, gauss_y)
plt.show()
```

```python
N = 10
rands = np.random.rand(100000, N)
x = np.sin(np.pi * (rands - 1 / 2))
data = (np.mean(x, axis=1)) / (np.sqrt(1 / 2) / np.sqrt(N))
plt.hist(data, bins=500, alpha=0.7, density=True)
plt.plot(lin_x, gauss_y)
plt.show()
```
